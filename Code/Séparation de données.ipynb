```python
selection=(range(187))
from numpy.random import permutation
permut = permutation(selection)
print(permut[:132])
ListDate=DF['DateTime'].unique()
print(ListDate[permut[:132]])
print(DF.loc[23]['DateTime'])
```

```python
ListDate=DF['DateTime'].unique()
ListeIndexTrain=[]
ListeIndexTest=[]
ListeIndexValid=[]

for i in range(DF.shape[0]):
  if DF.loc[i]['DateTime'] in ListDate[permut[:132]] :
    ListeIndexTrain.append(i)
  elif DF.loc[i]['DateTime'] in ListDate[permut[132:169]] :
    ListeIndexTest.append(i)
  elif DF.loc[i]['DateTime'] in ListDate[permut[169:]] :
    ListeIndexValid.append(i)
print(len(ListeIndexTrain)+len(ListeIndexTest)+len(ListeIndexValid))
```

```python
train_set=DF.loc[ListeIndexTrain][:].copy()
test_set=DF.loc[ListeIndexTest][:].copy()
valid_set=DF.loc[ListeIndexValid][:].copy()
```

```python
print(train_set.shape,test_set.shape,valid_set.shape)
train_set=train_set.drop('DateTime',axis=1)
test_set=test_set.drop('DateTime',axis=1)
valid_set=valid_set.drop('DateTime',axis=1)
print(train_set.shape,test_set.shape,valid_set.shape)
```
